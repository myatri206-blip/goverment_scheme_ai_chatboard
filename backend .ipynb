```python

def clean_text(text):

    text = text.lower()

    cleaned = ""
    for char in text:
        if char.isalnum() or char == " ":
            cleaned += char

        elif "\u0900" <= char <= "\u097F":
            cleaned += char
    return cleaned

df["clean_text"] = df["text"].apply(clean_text)

# Show before/after example
print("Before cleaning:", df["text"][29])
print("After cleaning: ", df["clean_text"][29])
print("\nText cleaning done!")
```

```python


X_train, X_test, y_train, y_test = train_test_split(
    df["clean_text"], df["intent"], test_size=0.2, random_state=42
)


vectorizer = TfidfVectorizer()
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)

# Logistic Regression model
model = LogisticRegression(max_iter=1000)
model.fit(X_train_vec, y_train)

# Test accuracy
y_pred = model.predict(X_test_vec)
accuracy = accuracy_score(y_test, y_pred)
print(f"Model trained successfully!")
print(f"Accuracy: {accuracy * 100:.1f}%")
```

```python

def predict_intent(text):
    """Takes English text, returns predicted category"""
    cleaned = clean_text(text)
    vector = vectorizer.transform([cleaned])
    return model.predict(vector)[0]


def translate_to_english(text):
    """Translates any language to English using deep-translator"""
    try:
        translated = GoogleTranslator(source='auto', target='en').translate(text)
        return translated
    except Exception as e:
        print(f"Translation note: {e}")
        return text


def chatbot(user_input):
    """Main chatbot function: returns HTML so links are clickable in Gradio"""

    if not user_input or user_input.strip() == "":
        return "<p>⚠️ कृपया अपना सवाल लिखें।</p>"

    # Step 1: Detect language
    try:
        lang = detect(user_input)
    except:
        lang = "en"

    # Step 2: Translate to English for ML model if pure Hindi
    if lang == "hi":
        english_text = translate_to_english(user_input)
    else:
        english_text = user_input

    # Step 3: Predict intent
    intent = predict_intent(english_text)

    # Step 4: Get scheme list
    scheme_list = schemes_hindi.get(intent, [])

    # Step 5: Build HTML response with clickable links
    # CHANGE 2: Output is now HTML instead of plain text
    # gr.HTML() renders this so links are actually clickable
    intent_labels = {
        "scholarship": "शिक्षा / छात्रवृत्ति",
        "agriculture": "खेती / किसान",
        "employment": "रोजगार / नौकरी",
        "women_support": "महिला सहायता",
        "health": "स्वास्थ्य / इलाज"
    }

    html = f"<h3>✅ आपकी जरूरत: {intent_labels.get(intent, intent)}</h3>"
    html += "<p>📋 <b>आपके लिए सरकारी योजनाएं:</b></p>"

    for name, eligibility, link in scheme_list:
        html += "<div style='border:1px solid #ddd; border-radius:8px; padding:10px; margin:8px 0;'>"
        html += f"<b>{name}</b><br>"
        html += f"<span style='color:green;'>{eligibility}</span><br>"

        if link:
            html += f"<a href='{link}' target='_blank' style='color:blue;'>🔗 हिंदी में वेबसाइट खोलें</a>"
        html += "</div>"


    html += "<hr><p>💡 नज़दीकी CSC केंद्र या ग्राम पंचायत में भी जानकारी मिलेगी।</p>"
    return html


print("✅ Chatbot function ready!")
```

```python

import speech_recognition as sr

def speech_to_text_from_file(audio_file_path):
    """
    Converts a .wav audio file to Hindi text.
    Use this in Colab by uploading a .wav file.
    """
    recognizer = sr.Recognizer()
    with sr.AudioFile(audio_file_path) as source:
        audio_data = recognizer.record(source)
    try:
        # language='hi-IN' means it will understand Hindi speech
        text = recognizer.recognize_google(audio_data, language='hi-IN')
        print(f"🎤 आपने बोला: {text}")
        return text
    except sr.UnknownValueError:
        return "⚠️ आवाज़ समझ नहीं आई, कृपया दोबारा बोलें"
    except sr.RequestError:
        return "⚠️ इंटरनेट कनेक्शन की समस्या है"


def speech_to_text_live():
    """
    Live microphone input — works on LOCAL machine only (not Colab).
    Run this on your laptop/desktop.
    """
    recognizer = sr.Recognizer()
    with sr.Microphone() as source:
        print("🎤 बोलिए... (Speak now...)")
        recognizer.adjust_for_ambient_noise(source, duration=1)
        audio_data = recognizer.listen(source, timeout=10)
    try:
        text = recognizer.recognize_google(audio_data, language='hi-IN')
        print(f"✅ आपने बोला: {text}")
        return text
    except sr.UnknownValueError:
        return "⚠️ आवाज़ समझ नहीं आई"
    except sr.RequestError:
        return "⚠️ इंटरनेट कनेक्शन की समस्या है"


# HOW TO USE IN COLAB:
# 1. Upload a .wav audio file to Colab files panel (left sidebar)
# 2. Uncomment the lines below and change the filename

# audio_text = speech_to_text_from_file("/content/your_audio.wav")
# print(chatbot(audio_text))

print("✅ Speech recognition functions ready!")
print("ℹ️  To use: upload a .wav file and call speech_to_text_from_file('your_file.wav')")
```
